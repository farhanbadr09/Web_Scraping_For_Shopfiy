```python
import selenium
from selenium import webdriver
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup 
import pandas as pd
import csv
```

```python
driver = webdriver.Chrome(executable_path='driver/chromedriver.exe')
pro_titles = []
pro_images = []
pro_price = []
body_html = []
pro_option = [] 
counter = 0
for page in range(1,2):
    driver.get("https://www.stylekorean.com/brands/58/laneige"+str(page))
    html = driver.page_source
    soup = BeautifulSoup(html,'html')
    a = soup.find('div',class_ = 'productlist_skin')
    all_ = a.find_all('a', class_="sct_a",href=True)
    for i in all_:
        
        url = i['href']
        title = i.text.replace("\n", "")
        #Product Title
        pro_titles.append(title)
        #Load Driver 
        driver.get(url)
        try:
            driver.find_element(By.XPATH, '//*[@id="sit_ov"]/table[2]/tbody/tr/td[2]/span').click()
            html = driver.page_source
            Option_soup = BeautifulSoup(html,'html')
            a = Option_soup.find('ul', class_='select2-results__options')
            b = a.text.replace('\xa0', '')
            b = b.replace(' ', '')
            pro_option.append(b)
        except:
            pro_option.append("Null")
        pro_html = driver.page_source
        pro_soup = BeautifulSoup(pro_html,'html')
         #Product Image
        main_img = pro_soup.find('a',class_="popup_item_image")
        main_img = main_img.find('img')
        pro_images.append(main_img['src'])
        #Product Price
        price = pro_soup.find('table', class_="sit_ov_tbl")
        price = price.find('tbody')
        price = price.find('span')
        price = price.text
        price = price.replace(" USD", "")
        pro_price.append(price)
        
        #Product Description Text
        inner_pro_desc = '"<p>'
        desc = pro_soup.find('div', class_="is_contents") 
        desc = desc.text.replace("\n", "")
        inner_pro_desc += desc
        #Product Descption Images
#         details_img = []
        desc_img = pro_soup.find('div', class_="is_contents") 
        for img in desc_img.find_all('img'):
            img_tag = "<img src='"+img['src']+"'>"
            inner_pro_desc += img_tag
#             details_img.append(img_tag)
        inner_pro_desc += '</p>"'
        body_html.append(inner_pro_desc)
        
        print(counter)
        counter+=1
        
    

        
```
